```
# Import necessary libraries
import random

# Define meal options with corresponding ingredients
meals = {
    'Scrambled eggs with spinach and cheese': ['eggs', 'spinach', 'cheese'],
    'Greek yogurt with sliced banana and almonds': ['Greek yogurt', 'banana', 'almonds'],
    'Grilled chicken salad with avocado and almonds': ['chicken', 'salad', 'avocado', 'almonds'],
    'Salmon and spinach salad with olive oil dressing': ['salmon', 'spinach', 'salad', 'olive oil'],
    'Beef stir-fry with broccoli and cauliflower rice': ['beef', 'broccoli', 'cauliflower rice'],
    'Chicken and vegetable stir-fry with cauliflower rice': ['chicken', 'vegetables', 'cauliflower rice'],
    'Grilled chicken with roasted vegetables': ['chicken', 'vegetables'],
    'Salmon and roasted vegetables': ['salmon', 'vegetables']
}

# Define basic recipes for each meal
recipes = {
    'Scrambled eggs with spinach and cheese': 'Beat the eggs and cook in a non-stick pan. Add spinach and cheese.',
    'Greek yogurt with sliced banana and almonds': 'Mix Greek yogurt with sliced banana. Top with almonds.',
    'Grilled chicken salad with avocado and almonds': 'Grill the chicken and slice. Mix with salad, avocado, and almonds.',
    'Salmon and spinach salad with olive oil dressing': 'Grill the salmon and serve on a bed of spinach salad. Drizzle with olive oil.',
    'Beef stir-fry with broccoli and cauliflower rice': 'Stir-fry the beef and broccoli. Serve with cauliflower rice.',
    'Chicken and vegetable stir-fry with cauliflower rice': 'Stir-fry the chicken and vegetables. Serve with cauliflower rice.',
    'Grilled chicken with roasted vegetables': 'Grill the chicken and serve with roasted vegetables.',
    'Salmon and roasted vegetables': 'Bake the salmon and serve with roasted vegetables.'
}

breakfast_options = list(meals.keys())[:2]
lunch_options = list(meals.keys())[2:4]
dinner_options = list(meals.keys())[4:]

# Define meal plan
meal_plan = {}
for day in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']:
    meal_plan[day] = {}
    meal_plan[day]['breakfast'] = random.choice(breakfast_options)
    meal_plan[day]['lunch'] = random.choice(lunch_options)
    meal_plan[day]['dinner'] = random.choice(dinner_options)

# Define shopping list
shopping_list = {}
for day in meal_plan:
    for meal in meal_plan[day]:
        for ingredient in meals[meal_plan[day][meal]]:
            if ingredient not in shopping_list:
                shopping_list[ingredient] = 1
            else:
                shopping_list[ingredient] += 1

# Print meal plan, recipes and shopping list
print('Meal Plan:')
for day in meal_plan:
    print(day)
    for meal_time in ['breakfast', 'lunch', 'dinner']:
        meal = meal_plan[day][meal_time]
        print(f'{meal_time.capitalize()}: {meal}')
        print(f'Recipe: {recipes[meal]}')
    print()

print('Shopping List:')
for ingredient in shopping_list:
    print(ingredient, '-', shopping_list[ingredient])
```

```
import json
import requests

# Get the meals from the JSON file.
with open("meals.json") as f:
    meals = json.load(f)

# Create a list of ingredients for each meal.
ingredients = []
for meal in meals:
    ingredients.append(meal["ingredients"])

# Get the price of each ingredient from Wegmans.
prices = []
for ingredients in ingredients:
    for ingredient in ingredients:
        response = requests.get(f"https://www.wegmans.com/api/products/search?query={ingredient}")
        if response.status_code == 200:
            price = response.json()[0]["salePrice"]
            prices.append(price)

# Print the price of each meal.
for meal, price in zip(meals, prices):
    print(f"{meal['name']}: ${price}")
```
